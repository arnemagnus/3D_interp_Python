```python
import numpy as np
import cytricubic
```

```python
x = np.linspace(-1,1,250)
y = np.linspace(-1,1,400)
z = np.linspace(-1,1,400)

foo = np.sin(np.pi*np.meshgrid(x,y,z,indexing='ij')[0])
```

```python
bar = cytricubic.TricubicInterpolator(x,y,z,foo)
```

```python
ix = np.linspace(-1.1,1.1,250)
iy = np.array([bar.ev(x1=i,x2=0,x3=0,kx1=1,kx2=0,kx3=0) for i in ix])
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
plt.scatter(ix,iy,s=4,c='k')
plt.plot(ix,np.pi*np.cos(np.pi*ix),'--')
#plt.xlim(-0.1,0.1)
```

```python
x,dx = np.linspace(-1,1,100,retstep=True)
y = np.linspace(-1,1,250)
z = np.linspace(-1,1,300)

foo = np.sin(np.pi*np.meshgrid(x,y,z,indexing='ij')[0])
bar = cytricubic.TricubicInterpolator(x,y,z,foo)
```
